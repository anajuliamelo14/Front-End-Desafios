```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
DataFrame = pd.read_csv('https://raw.githubusercontent.com/sharmaroshan/Zoo-Dataset/refs/heads/master/zoo.csv')
```

```python
DataFrame_enum = pd.get_dummies(DataFrame)
DataFrame_enum
```

```python
dados = DataFrame_enum.drop(columns=['class_type'])
alvo = DataFrame_enum['class_type']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dados, alvo, test_size=0.10, random_state=42)
```

```python
model = tf.keras.models.Sequential([

  tf.keras.layers.Dense(70, activation=tf.nn.relu),
  tf.keras.layers.Dense(40, activation=tf.nn.relu),
  tf.keras.layers.Dense(8)
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python

x_train = x_train.astype('float32')
x_test = x_test.astype('float32')


y_train = y_train.astype('float32')
y_test = y_test.astype('float32')

model.fit(x_train, y_train, epochs=50)
```

```python
model.evaluate(x_test, y_test)
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

previsoes_prob = model.predict(x_test)
previsoes = np.argmax(previsoes_prob, axis=1)
matrix = confusion_matrix(y_test, previsoes)

print(matrix)
```
